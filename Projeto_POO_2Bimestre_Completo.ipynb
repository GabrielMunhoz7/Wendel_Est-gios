```python
import sqlite3

# Classe Principal - EntrevistaEstagio_gab
class EntrevistaEstagio_gab:
    def __init__(self, nome_candidato_gab, curso_gab, status_gab):
        self.__nome_candidato_gab = nome_candidato_gab
        self.curso_gab = curso_gab
        self.status_gab = status_gab

    def get_nome_candidato_gab(self):
        return self.__nome_candidato_gab

    def set_nome_candidato_gab(self, novo_nome_gab):
        self.__nome_candidato_gab = novo_nome_gab

    def verificar_status_gab(self):
        if self.status_gab != "Aprovado":
            self.status_gab = "Reprovado"
        print(f"Status verificado: {self.status_gab}")

    def exibir_gab(self):
        print(f"Nome do Candidato: {self.get_nome_candidato_gab()}")
        print(f"Curso: {self.curso_gab}")
        print(f"Status da Entrevista: {self.status_gab}")
        print("-" * 30)

    def alterar_status_gab(self, novo_status_gab):
        self.status_gab = novo_status_gab
        print(f"Status alterado para: {self.status_gab}")

    def atualizar_nome_gab(self, novo_nome_gab):
        self.set_nome_candidato_gab(novo_nome_gab)
        print(f"Nome atualizado para: {self.get_nome_candidato_gab()}")

    def inserir_no_banco(self, conn):
        cursor = conn.cursor()
        cursor.execute("INSERT INTO entrevistas (nome, curso, status) VALUES (?, ?, ?)", 
                       (self.get_nome_candidato_gab(), self.curso_gab, self.status_gab))
        conn.commit()

    @staticmethod
    def consultar_banco(conn):
        cursor = conn.cursor()
        cursor.execute("SELECT * FROM entrevistas")
        entrevistas = cursor.fetchall()
        for entrevista in entrevistas:
            print(f"ID: {entrevista[0]}, Nome: {entrevista[1]}, Curso: {entrevista[2]}, Status: {entrevista[3]}")
        print("-" * 30)

    @staticmethod
    def excluir_do_banco(conn, id_entrevista):
        cursor = conn.cursor()
        cursor.execute("DELETE FROM entrevistas WHERE id = ?", (id_entrevista,))
        conn.commit()

    @staticmethod
    def atualizar_status_banco(conn, id_entrevista, novo_status):
        cursor = conn.cursor()
        cursor.execute("UPDATE entrevistas SET status = ? WHERE id = ?", (novo_status, id_entrevista))
        conn.commit()

    @staticmethod
    def atualizar_nome_banco(conn, id_entrevista, novo_nome):
        cursor = conn.cursor()
        cursor.execute("UPDATE entrevistas SET nome = ? WHERE id = ?", (novo_nome, id_entrevista))
        conn.commit()

# Subclasse
class EntrevistaEstagioTecnico_gab(EntrevistaEstagio_gab):
    def __init__(self, nome_candidato_gab, curso_gab, status_gab, nivel_tecnico_gab):
        super().__init__(nome_candidato_gab, curso_gab, status_gab)
        self.nivel_tecnico_gab = nivel_tecnico_gab

    def exibir_nivel_tecnico_gab(self):
        print(f"N√≠vel T√©cnico: {self.nivel_tecnico_gab}")

    def exibir_gab(self):
        super().exibir_gab()
        print(f"N√≠vel T√©cnico: {self.nivel_tecnico_gab}")
        print("-" * 30)

# Cria√ß√£o do banco
def criar_banco():
    conn = sqlite3.connect('entrevistas.db')
    cursor = conn.cursor()
    cursor.execute('''CREATE TABLE IF NOT EXISTS entrevistas (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        nome TEXT,
        curso TEXT,
        status TEXT
    )''')
    conn.commit()
    return conn

# Menu interativo
def menu_interativo():
    conn = criar_banco()
    while True:
        print("\n==== MENU DE ENTREVISTAS ====")
        print("1. Adicionar nova entrevista")
        print("2. Listar entrevistas")
        print("3. Alterar nome do candidato")
        print("4. Alterar status da entrevista")
        print("5. Excluir entrevista")
        print("6. Sair")
        opcao = input("Escolha uma op√ß√£o: ")

        if opcao == '1':
            nome = input("Nome do candidato: ")
            curso = input("Curso: ")
            status = input("Status (Aprovado, Pendente, Reprovado): ")
            entrevista = EntrevistaEstagio_gab(nome, curso, status)
            entrevista.inserir_no_banco(conn)
            print("Entrevista adicionada com sucesso.")

        elif opcao == '2':
            print("\n--- Lista de Entrevistas ---")
            EntrevistaEstagio_gab.consultar_banco(conn)

        elif opcao == '3':
            id_entrevista = int(input("ID da entrevista para alterar nome: "))
            novo_nome = input("Novo nome do candidato: ")
            EntrevistaEstagio_gab.atualizar_nome_banco(conn, id_entrevista, novo_nome)
            print("Nome atualizado com sucesso.")

        elif opcao == '4':
            id_entrevista = int(input("ID da entrevista para alterar status: "))
            novo_status = input("Novo status (Aprovado, Pendente, Reprovado): ")
            EntrevistaEstagio_gab.atualizar_status_banco(conn, id_entrevista, novo_status)
            print("Status atualizado com sucesso.")

        elif opcao == '5':
            id_entrevista = int(input("ID da entrevista para excluir: "))
            EntrevistaEstagio_gab.excluir_do_banco(conn, id_entrevista)
            print("Entrevista exclu√≠da com sucesso.")

        elif opcao == '6':
            print("Encerrando programa...")
            conn.close()
            break

        else:
            print("Op√ß√£o inv√°lida. Tente novamente.")

# Executar menu
menu_interativo()

```
